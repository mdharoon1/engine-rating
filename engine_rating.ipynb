Deal with missing data

Drop Dataframe columns containing either 50% or more than 50% NaN values

```python
perc = 50.0
min_count =  int(((100-perc)/100)*df.shape[0] + 1)
df = df.dropna( axis=1, 
                thresh=min_count)
```

```python
df.isnull().sum()
```

```python
missing=df.isnull().sum()
missing=missing[missing>0] #Selecting the features which have null values present in them


missing=pd.DataFrame({'Feature':missing.index, 'missing_value_%':sorted(missing.values/len(df)*100)})
missing
```

We have 3 features still left with missing data with one approx 50% missing data. So we will drop one feature with approx 50% missing data and fill rest 2 columns data

```python
#Filing the missing values with most occurance value
df = df. fillna(df['engineTransmission_engineOil_cc_value_0'].value_counts().index[0])
```

```python
#Filing the missing values with most occurance value
df = df. fillna(df['engineTransmission_engineSound_cc_value_0'].value_counts().index[0])
```

```python
#droping one features which has approx 50% missing data
df.drop(['engineTransmission_engineSound_cc_value_1'],axis=1,inplace=True)
```

```python
print(df.isnull().sum())
print(df.shape)
```

# Using Pearson Correlation

The Pearson Correlation measures the linear dependence between two variables X and Y.

The resulting coefficient is a value between -1 and 1 inclusive, where:

1: Total positive linear correlation.
0: No linear correlation, the two variables most likely do not affect each other.
-1: Total negative linear correlation.

# Data Preprocessing

Normalising odometer_reading variable

```python
odometer_reading_mean1=df['odometer_reading'].mean()
odometer_reading_std1=df['odometer_reading'].std()
df['odometer_reading']=(df['odometer_reading']-odometer_reading_mean1)/(odometer_reading_std1)
```

```python
df.head()
```

```python
df.drop(['appointmentId','inspectionStartTime','year','month'],axis=1,inplace=True)
```

Encoding the target label using LabelEncoder

```python
le.fit(df['rating_engineTransmission'])
df['rating_engineTransmission']= le.transform(df['rating_engineTransmission'])
```

```python
df.head()
```

Separating the features and target variable from the dataframe

```python
x=df.drop('rating_engineTransmission',axis=1)
y=df['rating_engineTransmission']
```

Splitting the data into training set & test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=1)
```

# Model Building

```python
# Training the decision tree classifier
model.fit(x_train,y_train)
```

```python
# Predictions
y_pred_d = model.predict(x_test)
```

```python
model.score(x_test, y_test)  # Accuracy of model
```

```python
import numpy as np
```

```python
error_rate = []
for i in range(1,55):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    pred_kn = knn.predict(x_test)
    error_rate.append(np.mean(pred_kn != y_test))
```

```python
plt.figure(figsize=(12,8))
plt.plot(range(1,55),error_rate,color='b',linestyle='dashed',marker='o')
plt.xlabel('k')
plt.ylabel('error_rate')
```
